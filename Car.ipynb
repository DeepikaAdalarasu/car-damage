```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the trained YOLO model
model = YOLO("/content/runs/detect/train3/weights/best.pt")

def predict_and_draw_damaged_car(image_path, conf_threshold=0.25, iou_threshold=0.5):
    """
    Predicts damaged areas in a car image, draws bounding boxes, and merges overlapping boxes.

    Args:
        image_path (str): Path to the input image.
        conf_threshold (float): Confidence threshold for predictions.
        iou_threshold (float): IoU threshold for non-max suppression.
    """
    # Run YOLO model inference
    results = model(image_path, conf=conf_threshold, iou=iou_threshold)

    # Load the input image
    image = cv2.imread(image_path)

    # List to store bounding boxes
    boxes = []

    # Loop through the results to extract bounding boxes
    for result in results:
        for box in result.boxes:
            # Extract confidence and class
            confidence = box.conf[0].item()
            class_id = int(box.cls[0].item())

            # Only process boxes with class_id = 0 (damaged areas)
            if confidence > conf_threshold and class_id == 0:
                x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
                boxes.append([x1, y1, x2, y2, confidence])
                # Draw the individual bounding box
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                label = f"Damaged ({confidence:.2f})"
                cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Merge overlapping boxes
    if boxes:
        merged_boxes = merge_overlapping_boxes(boxes)
        for (x1, y1, x2, y2) in merged_boxes:
            # Draw merged boxes in blue
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            label = "Merged Damage"
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Display the final image with bounding boxes
    cv2_imshow(image)

def merge_overlapping_boxes(boxes, iou_threshold=0.2):
    """
    Merges overlapping bounding boxes based on IoU.

    Args:
        boxes (list): List of bounding boxes [x1, y1, x2, y2, confidence].
        iou_threshold (float): IoU threshold for merging.

    Returns:
        merged_boxes (list): List of merged bounding boxes [x1, y1, x2, y2].
    """
    boxes = sorted(boxes, key=lambda x: x[4], reverse=True)  # Sort by confidence
    merged_boxes = []
    while boxes:
        x1, y1, x2, y2, _ = boxes.pop(0)
        new_box = [x1, y1, x2, y2]
        to_remove = []

        for i, (xx1, yy1, xx2, yy2, _) in enumerate(boxes):
            if compute_iou(new_box, [xx1, yy1, xx2, yy2]) > iou_threshold:
                # Merge boxes
                new_box = [
                    min(new_box[0], xx1),
                    min(new_box[1], yy1),
                    max(new_box[2], xx2),
                    max(new_box[3], yy2),
                ]
                to_remove.append(i)

        # Remove merged boxes
        boxes = [b for i, b in enumerate(boxes) if i not in to_remove]
        merged_boxes.append(new_box[:4])  # Append only the coordinates

    return merged_boxes

def compute_iou(box1, box2):
    """
    Computes the Intersection over Union (IoU) between two bounding boxes.

    Args:
        box1, box2 (list): Bounding boxes [x1, y1, x2, y2].

    Returns:
        float: IoU value.
    """
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    # Calculate intersection area
    inter_area = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)

    # Calculate union area
    box1_area = (box1[2] - box1[0] + 1) * (box1[3] - box1[1] + 1)
    box2_area = (box2[2] - box2[0] + 1) * (box2[3] - box2[1] + 1)
    union_area = box1_area + box2_area - inter_area

    return inter_area / union_area if union_area > 0 else 0

# Input image path
image_path = "/content/drive/MyDrive/ExtractedCarDamage/damage car5.jpg"

# Run the prediction and visualization function
predict_and_draw_damaged_car(image_path)
```
